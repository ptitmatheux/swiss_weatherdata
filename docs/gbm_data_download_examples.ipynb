```python
def download_csv(url):
    """
    Downloads a CSV file from a given URL and returns a Pandas DataFrame.
    """
    try:
        # in production, please check if a local copy of the files already exists. If so, please
        # send the ETag of the local resource (that you got in the response when initially
        # requesting the resource) in an If-None-Match header. The server will only send the file,
        # if the remote version is newer than your local file. This avoids unnecessary traffic.
        # (also check here: https://data.geo.admin.ch/api/stac/static/spec/v1/apitransactional.html#tag/Data/operation/getAssetObject)
        # For our short example, we don't need to do 4all this.
        df = pd.read_csv(url, delimiter=';')
        return df
    # too broad exception, I know. Please use better error handling in production ;-)
    except Exception as e:
        print(f"Error downloading {url}: {e}")
        return None
```

```python
def sanitize_column_names(df):
    """
    Removes single quotes from column names in a given Pandas DataFrame.
    Also rename "REFERENCE_TS" to "reference_timestamp".

    Args:
        df (pd.DataFrame): The input DataFrame.

    Returns:
        pd.DataFrame: A new DataFrame with cleaned column names.
    """
    new_column_names = {col: col.replace("'", "") for col in df.columns}
    df_cleaned = df.rename(columns=new_column_names)

    if "REFERENCE_TS" in df_cleaned.columns:
        df_cleaned = df_cleaned.rename(columns={"REFERENCE_TS": "reference_timestamp"})

    # if not globrad_col in df_cleaned.columns:
    #     print("Error: No valid global radiation column found!")
    #     return None


    return df_cleaned
```

```python
###### possibly adapt this:
BATCH_SIZE = 1e4
UPDATE_DT = True
#df_upload = df.head(1000) # this is for testing with a small number of records
#df_upload = df.iloc[0: int(1e6)] 
#df_upload = df.iloc[int(1.6e6):]
df_upload = df_reshaped
######
print(f'This is the head of the list of records that will be uploaded to the database: \n {df_upload.head()}')
print(f'This is the tail of the list of records that will be uploaded to the database: \n {df_upload.tail()}')

nb_of_batches = ceil(len(df_upload) / BATCH_SIZE)
print(f"splitting input data ({len(df_upload)} rows) into {nb_of_batches} batches")
for b in range(nb_of_batches):
    beg_slice = int(b * BATCH_SIZE)
    end_slice = int(min([(b + 1) * BATCH_SIZE, len(df_upload)]))
    print(f"batch {b+1}: processing slice from index {beg_slice} to index {end_slice - 1}")
    sub_df = df_upload.iloc[beg_slice:end_slice]
    data_dict = sub_df.to_dict(orient='records')
    new_values_list = [] 
    update_values_list = []
    with Session(engine) as session:
        session.begin()
        try:
            for row in data_dict:
                exist_stmt = and_(table_measures.c.nat_abbr_tx == row['nat_abbr_tx'],
                            table_measures.c.parameter_mch == row['parameter_mch'],
                            table_measures.c.time == row['time'])
                #print(exist_stmt)
                entry_exist = session.query(exists([table_measures]).where(exist_stmt)).scalar()
                #print(entry_exist)
                if entry_exist:
                    if UPDATE_DT:
                        update_stmt = update(table_measures).where(exist_stmt).values(value=row['value'])
                        update_results = session.execute(update_stmt)
                        update_values_list.append(row)
                    else:
                        pass
                else:
                    new_values_list.append(row)
                
            # inserting the new values in one batch:
            if len(new_values_list) > 0:
                results = session.execute(insert(table_measures), new_values_list)
            
            if len(update_values_list) > 0:
                print('The following already existing entries will be updated:')
                for i in range(len(update_values_list)):
                    print(update_values_list[i])
        except:
            session.rollback()
            raise
        else:
            session.commit()
            print('Number of new entries: ' + str(len(new_values_list)))
            print('Number of updated entries: ' + str(len(update_values_list)))
```
